# Numpy converstion


```python
import numpy as np
arr = np.array([1, 2, 3])
y = arr.astype(float)
print(y)
print("The type of y will be :",type(y))
```

```python
arr = np.array([1.7, 2.9, 3.1])
y = arr.astype(int) # When you do astype(int) on a float array, NumPy simply truncates (drops) the decimals — it doesn’t round.
print(y)
```

```python
arr = np.array(["1", "2", "3"])
y = arr.astype(int)
y
```

```python
arr = np.array(["1", "two", "3"])
y = arr.astype(int)
```

```python
arr = np.array([1, 2, 3])
y = arr.astype(bool)
print(y) # Every nonzero number becomes True, so [1, 2, 3] → [True, True, True].
```

```python
import numpy as np
arr = np.array([1, 2.5, 3]) # can’t mix int and float, it promotes them to the “wider” type.
print(type(arr))
```

```python
arr = np.array([1, "2", 3]) # here np will convert every thing into str but the dtype will shoe np.array
type(arr)
```

# Type converstion in Pandas 

```python
import pandas as pd
df = pd.DataFrame({
    'A': ['1', '2', '3'],
    'B': [1.5, 2.5, 3.5]
})
df['A'] = df['A'].astype(int)
print(type(df['A']))
```

```python
# Other ways to change dtypes:
df['A'] = pd.to_numeric(df['A'])
```
