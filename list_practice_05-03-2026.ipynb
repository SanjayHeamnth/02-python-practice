Accessing  list items in python


```python
a = [10, 20, 30, 40, 50]
print(a[0])
```

```python
# using list compration
d = [x for x in a ]
d
```

```python
# using list slcing
a[0:3]
```

```python
# using loop
for i in a :
    print(i)
```

Change list item


```python
a = [10, 20, 30, 40, 50]
a[1] = 22
a
```

```python
# useing slicing for multipal items
a[0:1] = [13,33]
```

```python
a
```

```python
#using list comprahention
f = [x**2 if x%2==0 else x for x in a if x>13]
f
```

```python
# using list with enumarate
for i ,x in enumerate(a):
    if x%2!=0:
        a[i]+=5
print(a)
```
